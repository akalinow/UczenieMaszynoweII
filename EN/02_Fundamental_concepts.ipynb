**"Machine" approach** - we define a set of functions, numbered by parameters $\theta$:
    <br></br>
    $$
    \huge{
     f(X, \theta)
    }
    $$
    <br></br>
    which may contain the function that solves our problem:
    <br></br>
    $$
    \huge{
     f(X, \hat{\theta}) = Y
    }
    $$
    <br></br>

We have to somehow define the function $f(X, \hat{\theta})$. We define it as a series expansion in some basis composed of functions $g_{i}$
    $$
    \huge{
     f(X, \hat{\theta}) = \sum_{i} \theta_{i} g_{i}(X)
    }
    $$

and look for the coefficients $\theta_{i}$    

Taylor's theorem states that any (reasonable) function can be expanded as an infinite sum of polynomials: 
$$
\huge{
\begin{matrix}
f(h, a_{n}) = 
\sum_{n=0}^{\infty} a_{n}h^{n} = \\
a_{0} + a_{1}h + a_{2}h^{2} +...,~~ h = x-x_{0}
\end{matrix}
}
$$
We can use them as our basis. The parameters $\theta_{i}$ are just the parameters of our series expansion:
<br></br>
$$
\huge{
a_{i} = \frac{f^{(n)}(x=x_{0})}{n!}
}
$$

If we know the coefficients, we are done! - **we know the $f(x)$**

What should we do if we don't know the series expansion, because we don't know the function? We only know the values at certain points, that is data as a set of pairs: $(x_{i}, y_{i})$
and we're looking for the relationship: $y_{i} = f(x_{i})$.

In this case we are looking for the `best` parameters according to some criterion, usually called the **loss function**
, $L$. 

The simplest criterion is the **mean square error(MSE)** :
<br></br>
$$
{\huge
MSE(\theta) = \frac{1}{n}\sum_{i} (y_{i} - f(x_{i}, \theta))^{2}
}
$$

**Few words about the notation:**

$x_{i}$ is the data for one point, example, in the input data. If the input data is one-dimensional it is one number, if it is k-dimensional there is k numbers.
A single example is thus a vector of numbers: $\mathbf{x} = (x_{0}, x_{1}, ...)$.
Multiple examples can be stacked sequentially, leading to a matrix representation $\mathbf{X}$ with n rows:

* **rows** - subsequent points, examples of data
* **kolumny** - subsequent features of the data point

$$
{\huge
X = \begin{bmatrix}
x_{0,0} & x_{1,1} & \dots & x_{0,k-1}\\
x_{1,0} & x_{1,1} & \dots & x_{1,,k-1}\\
\dots   & \dots   & \dots & \dots     \\ 
x_{k-1,0} & x_{k-1,1} & \dots & x_{n-1,,k-1}\\
\end{bmatrix}
}
$$

Similarly, values at different point are represented by a matrix $\mathbf{Y}$ with k rows:

* **rows** - values for consecutive data points
* **columns** - subsequent entries for the value if it is not a single number

In numpy notation we have an array of shape `(n,k)` and can easily select parts (slices) of it:

In special cases, e.g. when $f(x)$ is a linear function:

$$
{\large
f(\mathbf{X}, \mathbf{\theta}) = X \cdot \mathbf{\theta} + \theta_{0}
}
$$

The parameters minimizing the MSE can be found analytically. In other cases we need to look for a numerical solution.

Thins get complicated if we assume (usually right), that there is not a single deterministic relationship between $y$ and $x$, but the relationship is merely statistical.
The simplest model of such non-deterministic relationship is the presence of noise $\epsilon$ i $y$:
$$
\large{
y = f(x) + \epsilon
}
$$

 where $x$ and $y$ are now **random variables** described by a joint probability distribution $p(x,y)$

In that scenario the value of the loss function is an estimate of the expected value $p(x,y)$:

$$
{\large
\begin{matrix}
MSE = E_{x,y \sim p(x,y)} ( [Y - f(X, \theta)]^{2}) = \\
\int (Y - f(X, \theta))^{2} p(x,y) dx dy =
 \vert p(x,y) = p(y|x)p(x) \vert \\
\int \left[\int (Y - f(X, \theta))^{2}  p(y|x)p(x) dy \right] dx = \\
\int \left[\int (Y - f(X, \theta))^{2}  p(y|x) dy \right] p(x) dx = \\
E_{x \sim p(x)} \left[ E_{y \sim p(y|x)} (Y - f(X, \theta))^{2} \right]
\end{matrix}
}
$$


If we are looking for $f(x, \theta)$ that minimizes the MSE, it is sufficient to minimize for each of the `x` values separately:

$$
{\huge
f(x) = argmin_{c} E_{y \sim p(y|x)}((y-c)^{2})
}
$$

for a quadratic loss function the solution is:

$$
{\huge
f(x) = E_{y \sim p(y|x)}(y) = E(y|x)
}
$$

and for an absolute value loss function: 

$$
{\huge
f(x) = argmin_{c} E_{y \sim p(y|x)}(|y-c|)
}
$$

the solution is the median: 

$$
{\huge
f(x) = median(y|x)
}
$$

**Comment about the notation**:

* a problem of finding a real-valued function is called **regression**

* absolute value loss: 

$$
{\huge
|f(X) - Y|
}
$$

is called $\mathbf{L_{1}}$ loss

* quadratic loss of the form: 

$$
{\huge
(f(X) - Y)^{2}
}
$$

is called $\mathbf{L_{2}}$ loss.

* the averaging factor: $1/n$ is not always taken into account

When looking for a function that is not real-valued, but returns a category, `g`, where an example described by a vector of features $\mathbf{x} = (x_{0}, x_{1}, ...)$ belongs, and $g$ belongs to a finite set of categories $g \in G$ we are dealing with  **classification**.

In that case we can use a loss of `0`, `1` depending on `f(x)` correctly classifying the example based on the set of features.
This is called a **Bayes classifier**: 

$$
{\huge
f(x) = max_{g \in G} P(g|x)
}
$$

that assigns an example to a category with the highest probability given the observed set of features.
The `0-1` loss is non-dififerentiable, as it only changes in one point, so it is not numerically useful. 
Strata `0-1` jest jednak funkcjÄ… nierÃ³Å¼niczkowalnÄ…, ktÃ³re zmienia sieÂ tylko w jednym punkcie, wiÄ™Ä‡ nie jest uÅ¼yteczna numerycznie.

In real applications the function assigning categories is found in 2 steps:


1) $f(x)$ representing $p(g|x)$. It is anologous to regression - searching for a function with a continuum of values. We don't know the true $p(g|x)$, thus we cannot use the $L_{2}$ loss. Here, we use a loss function that leads to $f(x)$ which represents a probability distribution maximized by the training dataset:
 

$$
{\large
\begin{matrix}[l]
L(f(x)) = - \frac{1}{n}\sum_{i}\log f_{g}(x_{i}), \\
\\
\text{where:}\\
i~\text{- example index} \\
\text{g - true example category}\\
f_{g} - \text{probability estimate for} \mathbf{~true~} \text{category} \\
\end{matrix}
}
$$

2) selecting the proper class based on the estimate $p(g|x)$ (not necessarily using the Bayes criterion: $g: \text{argmax}_{g} p(g|x)$)


The loss $L(f(x))$ is called **cross-entropy**.

### Training

How to find the parameters of $\mathbf{\theta}$?

Calculate the gradient of the loss function with respect to the model parameters and modify the parameters in the direction opposite to the gradient.

Stochastic gradient descent, SGD:

1) initialize model parameters `init_theta`

2) select a batch of data from the training dataset.
   The loss function and gradient values will be averaged over batches to increase stability.

4) compute the loss function gradient **with respect to the model parameters**.
   For MSE, we have:
<br></br>
$$
\huge{
\nabla_{\theta} L = \nabla_{\theta} \frac{1}{N} \sum_{X} (y - f(x,\theta))^{2} = \\
\frac{2}{N} \sum_{X} (y - f(x,\theta)) \nabla_{\theta} f(x, \theta)
}
$$
<br></br>

4) update the parameter vector by subtracting the gradient scaled by the **learning rate**
<br></br>
$$
\huge{
 \mathbf{\theta_{i+1}} = \mathbf{\theta_{i}} - \alpha \cdot \nabla_{\theta} L
}
$$
<br></br>
5) go through the entire training dataset. One go is called an **epoch**

6) repeat the procedure until convergence criteria are met

<br></br>

In real applications we use advanced optimization algorithms.
One of the most popular is Adam `adaptive moment estimation` - [adam](https://arxiv.org/abs/1412.6980)

Each of the optimization algorithms uses the function gradient w.r.t. model parameters. 
Computing the gradient in complex models becomes computationally expensive. The gradient is computed using 
[automatic differentiation](https://en.wikipedia.org/wiki/Automatic_differentiation). This operation requires keeping track of the intermediate steps when computing a compound function, e.g. the loss function. 

The intermediate steps are stored on the tape when the value of $f(x)$ is computed.

* the process of computing the values from the input data: $x$ to the result $f(x)$ is called the forward pass, or feedforward

* the process of using the gradients to update the weights is called the backward pass, or feedbackward, or backpropagation
  
* the class responsible for storing the data required to compute the gradient is `tf.GradientTape`:

````python
x = tf.Variable(3.0)

with tf.GradientTape() as tape:
  y = x**2

print(colored("f(x) = x**2","blue"))  
print(colored("df/dx at x=3:","blue"),tape.gradient(y, x))  
````

Let us now find the values of the quadratic equation that best describes the data.
Let us implement the gradient descent algorithm using automatic differentiation. 

**Data:**

random numbers satisfying the equation

$$
\huge{
  -2 + 3x + 2x^{2}
}
$$

**Model:** 

$$
\huge{
  p_{0} + p_{1}x + p_{2}x^{2}
}
$$

**Model parameters:**

$$
\huge{
  p_{0}, p_{1}, p_{2}
}
$$

**Attention:** whenever we use TensorFlow library, we should use the functions from the library, instead of the corresponding numpy functions to prevent data type mismatch.

* TensorFlow uses `tf.Tensor` for elementary matrix representation:
  <br></br>
  ```Python
  x = tf.Variable([np.pi,np.exp(1)])
  print(x)
  <tf.Variable 'Variable:0' shape=(2,) dtype=float32, numpy=array([3.1415927, 2.7182817], dtype=float32)>
  ```
  <br></br>
* Numpy uses `np.array` for elementary matrix representation
  <br></br>
  ```Python
  x = np.array([np.pi,np.exp(1)])
  print(x)
  [3.14159265 2.71828183]
  ```

````python
%%time

#Input data - features
x = tf.random.uniform((1000,1), 0,1)

#input data - targets
y = 2*x**2 + 3*x -2

#model parameters - weights
weights = [tf.Variable(0.0, name="p2"),
           tf.Variable(0.0, name="p1"),
           tf.Variable(0.0, name="p0")]

print(colored("Initial model weights:","blue"))
[print(w.numpy(),end=" ") for w in weights]
print("")

#number of iterations - updates of the model parameters
nIterations = 5000

#learning rate
lr = 0.01

#the training loop
for iter in range(0,nIterations):
    #gradient tape preparation
    with tf.GradientTape() as tape:
            y_pred = tf.math.polyval(coeffs=weights, x=x)
            loss = tf.keras.losses.MeanSquaredError()(y_true = y, y_pred = y_pred)
            gradient = tape.gradient(loss, weights)
    #parameters update    
    for w,g in zip(weights, gradient):
        w.assign_sub(lr*g)
print(colored("Final model weights:","blue"))
[print(w.numpy(), end=" ") for w in weights]
print("")
````

````python
%%time

#Prepare the z dataset
nx = 2000
ny = 2000
r = 5

x = np.random.default_rng().uniform(-10,10,nx)
y = np.random.default_rng().uniform(-10,10,ny)
xv, yv = np.meshgrid(x, y)

#use ineffective method to show the numpy filtering
#abilities
circleFilter = np.abs(xv**2 + yv**2 - r**2)<0.001
xv = xv[circleFilter]
yv = yv[circleFilter]
z = np.array((xv,yv)).transpose()

#BEGIN_SOLUTION
#########
#define loss function as a mean distance of a proposed point from
#points in z dataset
def lossFunction(p, z):
    d = tf.subtract(z,p)
    d = tf.sqrt(tf.math.reduce_sum(d**2, axis=1))
    d = tf.math.reduce_mean(d)
    #d = tf.math.reduce_max(d)
    return d
#########

##Perform training loop
weights = tf.Variable((10.0, 10.0), name="p")

nIterations = 500
lr = 0.5

loss_history = []
gradient_history = np.array([[0,0]])
p_history = weights.numpy().reshape(-1,2)
#loop over oterations
for iter in range(0,nIterations):
    #gradient tape preparation
    with tf.GradientTape() as tape:
            loss = lossFunction(weights, z)
            gradient = tape.gradient(loss, weights)
            loss_history.append(loss)
            p_history = np.concatenate((p_history,weights.numpy().reshape(-1,2)), axis=0)
            gradient_history = np.concatenate((gradient_history,gradient.numpy().reshape(-1,2)), axis=0)
            weights.assign_sub(lr*gradient)
####      
print(colored("p = ","blue"),"({:3.3f}, {:3.3f})".format(weights[0].numpy(), weights[1].numpy()))
print(colored("Loss:","blue"),loss.numpy())
fig, axes = plt.subplots(1,2, figsize=(8,4))
axes[0].plot(xv, yv, marker='o', linestyle='none');
axes[0].plot(p_history[:,0],p_history[:,1], marker='o', linestyle='none');
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_xlim( (-10,10))
axes[0].set_ylim( (-10,10));

axes[1].plot(loss_history,label="loss")
axes[1].plot(gradient_history,label="gradient")
axes[1].set_xlabel('iteration')
axes[1].set_ylabel('')
axes[1].set_yscale('log')
axes[1].legend()
plt.subplots_adjust(bottom=0.15, left=0.05, right=0.95, wspace=0.3)
#END_SOLUTION
pass
````

* draw the set of points on a figure with axes in the range $\pm 12$
* draw the trajectory of a selected point during the minimization process
* on the same drawing include the analytical solution to the problem

**Attention:** the class `GradientTape` requires that all computation within the function for which the gradient is to be found are performed using TensorFlor functions. The parameters should be passed as `tf.Variable`

````python
%%time

#BEGIN_SOLUTION
#########
#define loss function as a mean distance of a proposed point from
#points in z dataset
def lossFunction(f1, f2, a, z):
    
    d1 = tf.subtract(z,f1)
    d2 = tf.subtract(z,f2)
    d1 = tf.sqrt(tf.math.reduce_sum(d1**2, axis=1))
    d2 = tf.sqrt(tf.math.reduce_sum(d2**2, axis=1))
    
    d = tf.add(d1, d2)
    d = tf.subtract(d,2*a)
    d = tf.abs(d)
    #d = tf.math.reduce_mean(d)
    d = tf.math.reduce_max(d)
    return d
#########
#define variables uses for loss calculation
f1 = tf.Variable([[-5.0, 0.0]])
f2 = tf.Variable([[5.0, 0.0]])
a = tf.Variable(10.0)

#define other ellipse parameters and calculate points on
#ellipse
epsilon = 0.5
c = epsilon*a
b = np.sqrt(a**2 - c**2)
nx = 100
x = np.linspace(-10,10,100)
y = b*np.sqrt(1 - (x/a)**2)

##perform the training loop
nPoints = 100
weights = tf.Variable(tf.random.uniform((nPoints, 2), minval=-10, maxval=10),name='w')

nIterations = 2000
lr = 0.1

loss_history = []
p_history = weights[0].numpy().reshape(-1,2)
#loop over oterations
for iter in range(0,nIterations):
    #gradient tape preparation
    with tf.GradientTape() as tape:
            loss = lossFunction(f1, f2, a, weights)
            loss_history.append(loss)
            p_history = np.concatenate((p_history,weights[0,:].numpy().reshape(-1,2)), axis=0)
            gradient = tape.gradient(loss, weights)
            weights.assign_sub(lr*gradient)

#### 
print(colored("Loss:","blue"),loss.numpy())
fig, axes = plt.subplots(1,2, figsize=(8,4))
axes[0].plot(p_history[:,0],p_history[:,1], marker='o', linestyle='none');
axes[0].plot(weights[:,0], weights[:,1], marker='o', linestyle='none');
axes[0].plot(x,y, color="black",linewidth=1);
axes[0].plot(x,-y, color="black",linewidth=1);
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_xlim( (-12,12))
axes[0].set_ylim( (-12,12));

axes[1].plot(loss_history,label="loss")
axes[1].set_xlabel('iteration')
axes[1].set_ylabel('')
axes[1].set_yscale('log')
axes[1].legend()
plt.subplots_adjust(bottom=0.15, left=0.05, right=0.95, wspace=0.3)
#END_SOLUTION
pass
````
